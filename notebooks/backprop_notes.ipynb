### Computing Derivatives with a Computation Graph



Let:
- `a = 5`
- `b = 3`
- `c = 2`

We define:
- `u = b * c`
- `v = a + u`
- `j = 3 * v`

Calculations:
- `u = b * c = 3 * 2 = 6`
- `v = a + u = 5 + 6 = 11`
- `j = 3 * v = 3 * 11 = 33`

#### Derivative of `j` with respect to `v` (dv):

The derivative of `j` with respect to `v` is 3. This means that if we change `v` by 0.001, `j` will increase 3 times more.

#### Finding `da` (how `a` affects `j`):

- Increasing `a` affects `v`, which then affects `j` (Chain Rule).
- To find this derivative, we multiply the rate of change of `v` with respect to `a` by `dv`.

Since `dv/da = 1` and `dv = 3`, we have:
- `da = 1 * 3 = 3`

#### Derivative of `j` with respect to `u` (how `u` affects `j`):

- The derivative of `j` with respect to `u` is 3.

#### Derivatives with respect to `b` and `c`:

- For `b`:
  - `du/db = c = 2` and `du = 3`
  - So, `dj/db = 3 * 2 = 6`

- For `c`:
  - `du/dc = b = 3` and `du = 3`
  - So, `dj/dc = 3 * 3 = 9`
